```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
def load_data(filepath):
    df=pd.read_csv(filepath)
    return df
def plot_histogram(df,column,bins=10):
    plt.figure(figsize=(8,6))
    sns.histplot(df[column],bins=bins,kde=True,color='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Wickets')
    plt.show()
def plot_scatter(df,x_column,y_column):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[x_column],y=df[y_column],color='red')
    plt.title(f'Scatter Plot of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
def plot_bar_chart(df,x_column,y_column):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=df[x_column],hue=df[x_column],legend=False,y=df[y_column],palette='viridis')
    plt.title(f'Bar Chart of {x_column} by {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.show()
file_path="/content/cricket_dataset-4.csv"
data=load_data(file_path)
plot_histogram(data,'Wickets')
plot_scatter(data,'Runs','Matches')
plot_bar_chart(data,'100s','Runs')
```
