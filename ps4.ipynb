```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

N = 1000

customer_counts = np.arange(0, 11)
probabilities = np.array([0.05, 0.08, 0.12, 0.15, 0.10, 0.07, 0.05, 0.03, 0.02, 0.01, 0.01])

probabilities = probabilities / probabilities.sum()

customer_data = np.random.choice(customer_counts, size=N, p=probabilities)

unique, counts = np.unique(customer_data, return_counts=True)
pmf = counts / N

cdf = np.cumsum(pmf)

plt.figure(figsize=(10, 5))
sns.barplot(x=unique, y=pmf, color='skyblue', edgecolor='black')
plt.xlabel("Number of customers per day")
plt.ylabel("Probability(PMF)")
plt.title("Probability Mass Function (PMF) of Customer Data")
plt.xticks(unique)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(x=unique, y=cdf, marker='o',color='skyblue',linestyle='-')
plt.xlabel("Number of customers per day")
plt.ylabel("Cumulative Probability(CDF)")
plt.title("Cumulative Distribution Function (CDF) of Customer Visits")
plt.xticks(unique)
plt.ylim(0,1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import scipy.stats as stats
N=100
p=0.70
prob_60_pass=stats.binom.pmf(60,N,p)
prob_at_least_75=1-stats.binom.cdf(74,N,p)
expected_pass=N*p
print(f"Probability of exactly 60 students passing: {prob_60_pass:.4f}")
print(f"Probability of at least 75 students passing: {prob_at_least_75:.4f}")
print(f"Expected number of students passing: {expected_pass:.2f}")
```
