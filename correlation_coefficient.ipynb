```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {'Advertising Budget': [1000, 2000, 3000, 4000, 1500, 2500, 3500, 1200, 4500, 5000],
        'Sales Revenue': [5000, 7000, 8000, 9500, 6000, 7500, 8700, 5500, 9900, 11000],
        'Social Media Engagement': [200, 250, 300, 400, 220, 280, 350, 210, 450, 500]}
df = pd.DataFrame(data)
```

```python
# Advertising Budget vs. Sales Revenue
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Advertising Budget', y='Sales Revenue', data=df)
plt.title('Advertising Budget vs. Sales Revenue')
plt.show()

correlation1 = df['Advertising Budget'].corr(df['Sales Revenue'])
print(f"Correlation between Advertising Budget and Sales Revenue: {correlation1}")
```

```python
# Social Media Engagement vs. Sales Revenue
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Social Media Engagement', y='Sales Revenue', data=df)
plt.title('Social Media Engagement vs. Sales Revenue')
plt.show()

correlation2 = df['Social Media Engagement'].corr(df['Sales Revenue'])
print(f"Correlation between Social Media Engagement and Sales Revenue: {correlation2}")
```

```python
# Advertising Budget vs. Social Media Engagement
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Advertising Budget', y='Social Media Engagement', data=df)
plt.title('Advertising Budget vs. Social Media Engagement')
plt.show()

correlation3 = df['Advertising Budget'].corr(df['Social Media Engagement'])
print(f"Correlation between Advertising Budget and Social Media Engagement: {correlation3}")
```
