```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

def plot_histogram(df, column, bins=10):
    plt.figure(figsize=(8,6))
    sns.histplot(df[column], bins=bins, kde=True, color='blue')
    plt.title(f'Histogarm of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

def plot_scatter(df,x_column,y_column):
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=df[x_column], y=df[y_column], color='red')
    plt.title(f'Scatter plot of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
def plot_bar_chart(df, x_column, y_column):
    plt.figure(figsize=(8, 6))
    sns.barplot(x=df[x_column], y=df[y_column], palette='viridis')
    plt.title(f'Bar Chart of {y_column} by {x_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.show()

file_path="/content/employee_data_250.csv"
data=load_data(file_path)

print(data.columns)

plot_histogram(data,'Salary')

# Changed 'Matches' to 'Age' and 'Wickets' to 'Salary' based on the available columns
plot_scatter(data,'Age','Salary')

# Changed 'Player' to 'Name' based on the available columns
plot_bar_chart(data,'Name','Salary')
```
