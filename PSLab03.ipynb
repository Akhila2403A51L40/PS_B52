```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
def load_data(filepath):
    df=pd.read_csv(filepath)
    return df
def plot_histogram(df,column,bins=10):
    plt.figure(figsize=(8,6))
    sns.histplot(df[column],bins=bins,kde=True,color='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
def plot_scatter(df,x_column,y_column):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[x_column],y=df[y_column],color='red')
    plt.title(f'Scatter Plot of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
def plot_bar_chart(df,x_column,y_column):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=df[x_column],hue=df[x_column],legend=False,y=df[y_column],palette='viridis')
    plt.title(f'Bar Chart of {x_column} by {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.show()
def plot_grouped_bar_chart(df, category_col, group_col):
    plt.figure(figsize=(12, 6))
    sns.countplot(x=df[category_col], hue=df[group_col], palette="magma")
    plt.title(f'{group_col} Distribution Across {category_col}')
    plt.xlabel(category_col)
    plt.ylabel("Number of Employees")
    plt.xticks(rotation=45)
    plt.legend(title=group_col)
    plt.show()
file_path="/content/employee_data_250-23a3c3fc-f06a-45e8-83af-27b87fa5b10a.csv"
data=load_data(file_path)
plot_histogram(data,'Salary')
plot_scatter(data,'Experience (Years)','Salary')
plot_bar_chart(data,'Experience (Years)','Department')
plot_grouped_bar_chart(data, 'Department', 'Job Satisfaction')
```
