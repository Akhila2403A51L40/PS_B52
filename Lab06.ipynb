Question 1:
Generate a normal distribution of data points with a mean of 75 and a standard
deviation of 10. Plot a histogram for the generated data. Analyze the symmetry and
bell-shaped curve of a normal distribution and visualize its Probability Density
Function (PDF)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

mean = 75
sigma = 10
N = 1000
np.random.seed(42)
data = np.random.normal(mean, sigma, N)

plt.figure(figsize=(10, 5))
sns.histplot(data, bins=30, kde=True, stat="density", color="skyblue", edgecolor="black", label="Histogram")
x = np.linspace(mean - 4*sigma, mean + 4*sigma, 1000)
pdf = norm.pdf(x, mean, sigma)
plt.plot(x, pdf, color="red", label="PDF (Normal Curve)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram and Probability Density Function (PDF)")
plt.legend()
plt.grid(True)
plt.show()
```

Question 2:
The normalized distribution for a persons body mass index (BMI) has a mean of 20
and a standard deviation of 4. Calculates and visualizes the following probabilities for
a randomly selected person having a BMI:
I. greater than or equal to 23.
II. less than or equal to 23.
III. between 23 and 35.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

mean_BMI = 20
std_dev_BMI = 4

p_greater_23 = 1 - norm.cdf(23, mean_BMI, std_dev_BMI)
p_less_23 = norm.cdf(23, mean_BMI, std_dev_BMI)
p_between_23_35 = norm.cdf(35, mean_BMI, std_dev_BMI) - norm.cdf(23, mean_BMI, std_dev_BMI)

print(f"Probability of BMI  23: {p_greater_23:.4f}")
print(f"Probability of BMI  23: {p_less_23:.4f}")
print(f"Probability of 23  BMI  35: {p_between_23_35:.4f}")

x = np.linspace(mean_BMI - 4*std_dev_BMI, mean_BMI + 4*std_dev_BMI, 1000)
pdf = norm.pdf(x, mean_BMI, std_dev_BMI)
plt.figure(figsize=(10, 5))
sns.lineplot(x=x, y=pdf, color="blue", label="BMI Distribution (Normal Curve)")
x_shade1 = np.linspace(23, mean_BMI + 4*std_dev_BMI, 100)
plt.fill_between(x_shade1, norm.pdf(x_shade1, mean_BMI, std_dev_BMI), color='red', alpha=0.4, label="BMI  23")
x_shade2 = np.linspace(mean_BMI - 4*std_dev_BMI, 23, 100)
plt.fill_between(x_shade2, norm.pdf(x_shade2, mean_BMI, std_dev_BMI), color='green', alpha=0.4, label="BMI  23")
x_shade3 = np.linspace(23, 35, 100)
plt.fill_between(x_shade3, norm.pdf(x_shade3, mean_BMI, std_dev_BMI), color='purple', alpha=0.4, label="23  BMI  35")
plt.xlabel("BMI")
plt.ylabel("Density")
plt.title("BMI Probability Distribution and Shaded Probabilities")
plt.legend()
plt.grid(True)
plt.show()
```
