Lab Exercise: To analyze and visualize an employee dataset to explore trends and
relationships among various factors using histograms, scatterplots, and bar charts.
Dataset Description: You are provided with a employee data csv file, which contains
250 observations with the following variables:
1. Employee ID – Unique identifier for each employee (Numeric).
2. Name – Name of the employee (Text).
3. Department – The department the employee works in (Categorical).
4. Age – Age of the employee in years (Numeric).
5. Salary – Monthly salary of the employee in USD (Numeric).
6. Experience – Years of experience the employee has (Numeric).
7. Job Satisfaction – Score on a scale of 1 to 5 (Numeric).
Tasks:
1. Histogram Analysis:
a. Plot a histogram to visualize the distribution of Salary across employees.
b. Use appropriate bin sizes and axis labels.
c. Identify the most frequent salary range.
2. Scatterplot Exploration:
a. Create a scatterplot of Experience vs. Salary to analyze the relationship between
work experience and earnings.
b. Use colors to differentiate job levels.
c. Add axis labels and a trendline to analyze trends.
3. Bar Chart Visualization:
a. Generate a bar chart showing the number of employees in each Department.
b. Identify which department has the highest number of employees.
c. Ensure the chart includes proper titles and axis labels.
4. Grouped Bar Chart:
a. Create a grouped bar chart comparing the number of employees by Job Level
across different Departments.
b. Discuss any visible trends regarding job level distribution within departments.
Analysis Questions:
1. What is the most common salary range in the dataset?
2. Does experience have a strong correlation with salary? Explain your findings.
3. Which department has the highest number of employees?
4. How is the distribution of job levels across different departments?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

def plot_histogram(df, column, bins=10):
    plt.figure(figsize=(8,6))
    sns.histplot(df[column], bins=bins, kde=True, color='blue')
    plt.title(f'Histogarm of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

def plot_scatter(df,x_column,y_column):
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=df[x_column], y=df[y_column], color='red')
    plt.title(f'Scatter plot of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
def plot_bar_chart(df, x_column, y_column):
    plt.figure(figsize=(8, 6))
    sns.barplot(x=df[x_column], y=df[y_column], palette='viridis')
    plt.title(f'Bar Chart of {y_column} by {x_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.show()

file_path="/content/employee_data_250.csv"
data=load_data(file_path)

print(data.columns)

plot_histogram(data,'Salary')

# Changed 'Matches' to 'Age' and 'Wickets' to 'Salary' based on the available columns
plot_scatter(data,'Age','Salary')

# Changed 'Player' to 'Name' based on the available columns
plot_bar_chart(data,'Name','Salary')
```
